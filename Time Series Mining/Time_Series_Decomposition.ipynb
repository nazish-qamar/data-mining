```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.dates import DateFormatter
from statsmodels.tsa.seasonal import seasonal_decompose
```

#### Dataset Source: https://community.qlik.com/t5/Qlik-Sense-Documents/AirPassengers-csv/ta-p/1476807

```python
df = pd.read_csv('./AirPassengers.csv', parse_dates=['Month'], index_col=['Month'])
```

#### Plot the data

```python
fig, ax = plt.subplots()
plt.xlabel('Year')
plt.ylabel('Monthly Passengers (1000s)')
plt.title('Monthly Airline Passengers')
plt.plot(df, color='black')
ax.xaxis.set_major_formatter(DateFormatter('%Y'))
```

#### Decompose the time series into trend, seasonal, and residuals.

```python
# Set the figure size
plt.rcParams['figure.figsize'] = [7, 8]

# Plot the decomposition components
sd = seasonal_decompose(df, period=12).plot()
```

- As we see growth over time, it would be appropriate to use a multiplicative trend.
- The residuals would be centered on 1 instead of 0.

```python
sd = seasonal_decompose(df, model='multiplicative').plot()
```
